## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import pandas as pd

csv_path = "/content/flyrank-ml-internship-starter/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(csv_path)
print("Rows, columns:", df.shape)

# 1) Sanity check: does position actually relate to CTR the way real search data should?
sub = df[(df["avg_position"] > 0) & (df["impressions_90d"] >= 100)]
corr_pos_ctr = sub["avg_position"].corr(sub["ctr"])
print(f"\n1) Correlation between avg_position and ctr (n={len(sub)}): {corr_pos_ctr:.2f}")

# 2) Does content age relate to decline rate?
df["is_declining"] = (df["trend_direction"] == "down").astype(int)
decline_by_age = df.groupby("age_tier")["is_declining"].mean().sort_values()
print("\n2) Decline rate by age_tier:")
print(decline_by_age)

# 3) Does time-since-last-update relate to decline rate?
decline_by_freshness = df.groupby("freshness_tier")["is_declining"].mean().sort_values()
print("\n3) Decline rate by freshness_tier:")
print(decline_by_freshness)
```
